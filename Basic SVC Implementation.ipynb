```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## lets create synthetic data points
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples =1000, n_features=2, 
                          n_classes=2, n_clusters_per_class=2, n_redundant=0) #for test i chagnes n_clusters_per_class to 2 you can put it 1 also
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)[0]
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1],hue=y)
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
# inport svc
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear')
```

```python
svc.fit(X_train,y_train)
```

```python
# prediction
y_pred = svc.predict(X_test)
```

```python
#performace metrics
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
svc.intercept_, svc.coef_
```

```python
#using rbf kernel
rbf=SVC(kernel='rbf')
rbf.fit(X_train,y_train)
y_pred1 = rbf.predict(X_test)
print(classification_report(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
```

```python
#using polynomial kernel
poly=SVC(kernel='poly')
poly.fit(X_train,y_train)
y_pred2 = poly.predict(X_test)
print(classification_report(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
```

```python
#using sigmoidal kernel
sig=SVC(kernel='sigmoid')
sig.fit(X_train,y_train)
y_pred3 = sig.predict(X_test)
print(classification_report(y_test,y_pred3))
print(confusion_matrix(y_test,y_pred3))
```

```python
sig.intercept_
```

### Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
 
# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}
```

```python
grid=GridSearchCV(SVC(),param_grid=param_grid,refit=True,cv=5,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
y_pred4 = grid.predict(X_test)
print(classification_report(y_test,y_pred4))
print(confusion_matrix(y_test,y_pred4))
```
